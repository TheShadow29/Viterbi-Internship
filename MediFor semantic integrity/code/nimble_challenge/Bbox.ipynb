```python
%matplotlib
import skimage.io as skio
import scipy.io as sio
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
```

```python
img = skio.imread('/home/nkovvuri/Rama_Work/dataset/Protest_Images/Modified_Images_ProtestL/beach_mdf_019.jpg')
can_box = sio.loadmat('/home/nkovvuri/Rama_Work/dataset/Protest_Images/Modified_Images_ProtestL/SS_boxes/beach_mdf_019.mat')
can_box2 = can_box['cur_bbxes'] - 1
```

```python
def show_bbx(img, bbx_list):
    # y1,x1, y2, x2
    fig, ax = plt.subplots(1)
    ax.imshow(img)

    for bbx in bbx_list:
        rect = mpatches.Rectangle((bbx[1], 960 -bbx[0]), bbx[3] - bbx[1], bbx[2] - bbx[0], edgecolor='red', facecolor='none')
        ax.add_patch(rect)
```

```python
show_bbx(img, can_box2[9:10])
```
